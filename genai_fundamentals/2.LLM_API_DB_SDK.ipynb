### Creating a Workspace Client

```python
# Import WorkspaceClient to interact with Databricks workspace APIs
from databricks.sdk import WorkspaceClient

# Import ChatMessage and ChatMessageRole to structure messages for LLM chat requests
from databricks.sdk.service.serving import ChatMessage, ChatMessageRole

# Create a client object to access Databricks workspace features and services
w = WorkspaceClient()
```

### Sending a Chat Completion Request to the LLM

```python
# Send a chat completion request to the Databricks LLM serving endpoint
response = w.serving_endpoints.query(
    name="databricks-gpt-oss-20b",  # Specify the LLM model to use
    messages = [
        # Provide a system message to set the context for the LLM
        ChatMessage(role=ChatMessageRole.SYSTEM, content="You are batman - the protector of Gotham City."),
        # Provide a user message with the question to ask the LLM
        ChatMessage(role=ChatMessageRole.USER, content="How's Gotham City doing today?")
    ],
    max_tokens = 1024  # Set the maximum number of tokens in the response
)

# Print the content of the LLM's response message
print(f"RESPONSE:{response.choices[0].message.content}")
```
