### Creating the OpenAI Client

```python
import os
from openai import OpenAI

client = OpenAI(
    api_key = os.environ.get("OPENAI_API_KEY"),  # Retrieve API key from environment variable
    base_url = "https://8259564857436390.0.gcp.databricks.com/serving-endpoints"
)
```

### Sending a simple chat request

```python
# Send a chat completion request to the OpenAI API using the specified model and messages
completion = client.chat.completions.create(
    model="databricks-gpt-oss-20b",  # Specify the model to use for generating the response
    messages=[
        {
            "role":"system",  # Set the role as 'system' to define the assistant's behavior
            "content": [
                {
                    "type": "text", "text": "You are Batman, the protector of Gotham City"  # Provide system instructions for the assistant
                }
            ]
        },
        {
            "role": "user",  # Set the role as 'user' to represent the user's input
            "content": [
                {
                    "type": "text", "text": "How's Gotham City doing today?"  # The user's message to the assistant
                }
            ]
        }
    ]
)

# Print the content of the assistant's response to the user's message
print(completion.choices[0].message.content)
```

### Sending Image as a Chat Request

![interrogation_scene](./Assets/interrogation_scene.png)

```python
import base64  # Import the base64 module for encoding binary data to base64 format
import httpx   # Import the httpx library for making HTTP requests

# Step 1: Download and encode the image
image1_url = "https://raw.githubusercontent.com/kuljotSB/DatabricksGenAIEngineer/refs/heads/main/GenAI_Fundamentals/Assets/interrogation_scene.png"  # URL of the image to be used
image1_data = base64.standard_b64encode(httpx.get(image1_url).content).decode("utf-8")  # Download the image, encode it in base64, and convert to string

# Step 2: Send a chat request to OpenAI with the image
completion = client.chat.completions.create(
    model = "databricks-llama-4-maverick",  # Specify the model to use for generating the response
    messages = [
        {
            "role": "system",  # System message to set the assistant's behavior
            "content": [
                {
                    "type": "text",
                    "text": "You are batman, the Dark Knight and Protector of Gotham City"  # Instruction for the assistant
                }
            ]
        },
        {
            "role": "user",  # User message containing the request and the image
            "content": [
                {
                    "type": "text",
                    "text": "Describe this image pls"  # User asks the assistant to describe the image
                },
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/png;base64,{image1_data}"  # Attach the base64-encoded image to the request
                    }
                }
            ]
        }
    ]
)

# Step 3: Print the assistant's response to the user's request
print(completion.choices[0].message.content)  # Display the description generated by the assistant
```
